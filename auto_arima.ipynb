```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima.arima import auto_arima
import sklearn
from arch import arch_model
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

### Load Data

```python
tickers = ['^GSPC', '^FTSE', '^N225', '^GDAXI']
start = "1994-01-07"
end = '2018-01-29'
raw_data = yfinance.download(tickers = tickers, start = start, end = end, interval = '1d')['Close']
```

```python
df_comp = raw_data.copy()
```

```python
df_comp.head()
```

```python
df_comp = df_comp.asfreq('b')
df_comp = df_comp.fillna(method='ffill')
```

### Create Returns

```python
ret = df_comp.pct_change(1)*100
ret = ret.iloc[1:,:]
ret.head()
```

### Split Data

```python
size = int(len(ret)*0.8)
train = ret.iloc[:size,:]
test = ret.iloc[size:,:]
```

### Fit AUTO ARIMA Model

```python
from pmdarima.arima import auto_arima
```

```python
auto = auto_arima(train['^GSPC'])
```

### Simple Forecasting

```python
# AR
model_ret_ar = ARIMA(train['^GSPC'], order = (1,0,0))
results_ret_ar = model_ret_ar.fit()
start_date = '2013-04-06'
end_date = '2015-01-01'
df_pred_ar = results_ret_ar.predict(start=start_date, end= end_date)

df_pred_ar[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (AR)', size=24)
plt.show()
```

```python
# MA
model_ret_ma = ARIMA(train['^GSPC'], order = (0,0,1))
results_ret_ma = model_ret_ma.fit()
df_pred_ma = results_ret_ma.predict(start=start_date, end= end_date)

df_pred_ma[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (MA)', size=24)
plt.show()
```

```python
# ARMA
model_ret_arma = ARIMA(train['^GSPC'], order = (1,0,1))
results_ret_arma = model_ret_arma.fit()
df_pred_arma = results_ret_arma.predict(start=start_date, end= end_date)

df_pred_arma[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (ARMA)', size=24)
plt.show()
```

```python
# ARMAX
model_ret_armax = ARIMA(train['^GSPC'], exog=train['^GDAXI'], order = (1,0,1))
results_ret_armax = model_ret_armax.fit()
df_pred_armax = results_ret_armax.predict(start=start_date, end= end_date, exog=test['^GDAXI'][start_date:end_date])

df_pred_armax[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (ARMAX)', size=24)
plt.show()
```

```python
# SARMA
model_ret_sarma = SARIMAX(train['^GSPC'], order = (1,0,1), seasonal_order = (3,0,2,5))
results_ret_sarma = model_ret_sarma.fit()
df_pred_sarma = results_ret_sarma.predict(start=start_date, end= end_date)

df_pred_sarma[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (SARMA)', size=24)
plt.show()
```

```python
# SARIMAX
model_ret_sarimax = SARIMAX(train['^GSPC'], order = (1,0,1), exog=train['^GDAXI'], seasonal_order = (3,0,2,5))
results_ret_sarimax = model_ret_sarimax.fit()
df_pred_sarimax = results_ret_sarimax.predict(start=start_date, end= end_date, exog=test['^GDAXI'][start_date:end_date])

df_pred_sarimax[start_date:end_date].plot(figsize = (20,5), color = 'red')
test['^GSPC'][start_date:end_date].plot(color = 'blue')
plt.title('Predictions vs Actual Returns (SARIMAX)', size=24)
plt.show()
```

### Auto ARIMA

```python
model_auto = auto_arima(train['^GSPC'], exogenous = train['^GDAXI'], m=5, max_p=5, max_P=5, max_Q=5)
```

```python
df_auto_pred = pd.DataFrame(model_auto.predict(n_periods = len(test[start_date:end_date]),
                            exogenous = test['^GDAXI'][start_date:end_date]),
                           index = test[start_date:end_date].index)
```

```python
ax = df_auto_pred.plot(figsize=(20, 5), color='red', legend=False)
test['^GSPC'][start_date:end_date].plot(
    ax=ax,
    color='blue',
    label='Actual'
)
ax.legend()
plt.title('Auto Model Predictions vs Real Data', size=24)
plt.show()
```

### Forecasting Volatility

```python
mod_garch= arch_model(ret['^GSPC'], vol = 'GARCH', p =1, q=1, mean='constant', dist='Normal')
res_garch = mod_garch.fit(last_obs = start_date, update_freq = 10)
```

```python
pred_garch = res_garch.forecast(horizon = 1, align = 'target')
```

```python
ax = pred_garch.residual_variance[start_date:end_date] \
    .plot(
        figsize=(20, 5),
        color='red',
        zorder=2,
        legend=False,
        label='Predicted Variance'
    )

test['^GSPC'][start_date:end_date].abs() \
    .plot(
        ax=ax,
        color='blue',
        zorder=1,
        label='Absolute Returns'
    )

ax.legend(fontsize=12)

ax.set_title('Volatility Predictions', size=24)
plt.show()
```
